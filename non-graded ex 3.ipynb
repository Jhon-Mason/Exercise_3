```python
import random

grid_size = 5
max_moves = 10

player_position = [random.randint(0, grid_size - 1), random.randint(0, grid_size - 1)]
treasure_position = [random.randint(0, grid_size - 1), random.randint(0, grid_size - 1)]

while player_position == treasure_position:
    treasure_position = [random.randint(0, grid_size - 1), random.randint(0, grid_size - 1)]

def calculate_distance(player_pos, treasure_pos):
    return abs(player_pos[0] - treasure_pos[0]) + abs(player_pos[1] - treasure_pos[1])

initial_distance = calculate_distance(player_position, treasure_position)

def print_grid(player_pos, treasure_pos, show_treasure=False, hide_player=False):
    for i in range(grid_size):
        row = []
        for j in range(grid_size):
            if [i, j] == player_pos and not hide_player:
                row.append('P')
            elif [i, j] == treasure_pos and show_treasure:
                row.append('T')
            else:
                row.append('.')
        print(" ".join(row))
    print()

print(f"Initial distance to treasure: {round(initial_distance, 2)}")
print("Initial grid:")
print_grid(player_position, treasure_position)

def make_move(player_pos, move):
    if move == 'N':
        if player_pos[0] > 0:
            player_pos[0] -= 1
        else:
            print("Move not allowed. You are at the northern edge of the grid.")
            return False
    elif move == 'S':
        if player_pos[0] < grid_size - 1:
            player_pos[0] += 1
        else:
            print("Move not allowed. You are at the southern edge of the grid.")
            return False
    elif move == 'E':
        if player_pos[1] < grid_size - 1:
            player_pos[1] += 1
        else:
            print("Move not allowed. You are at the eastern edge of the grid.")
            return False
    elif move == 'W':
        if player_pos[1] > 0:
            player_pos[1] -= 1
        else:
            print("Move not allowed. You are at the western edge of the grid.")
            return False
    else:
        print("Invalid move. Use 'N', 'S', 'E', or 'W'.")
        return False
    
    return True

moves_made = 0
while moves_made < max_moves:
    move = input("Enter your move (N, S, E, W): ").upper()
    prev_position = player_position.copy()

    move_successful = make_move(player_position, move)

    new_distance = calculate_distance(player_position, treasure_position)

    if not move_successful:
        print(f"New distance to treasure: {round(new_distance, 2)}")
    else:
        if player_position == treasure_position:
            print(f"Congratulations! You found the treasure!")
            print_grid(player_position, treasure_position, show_treasure=True, hide_player=True)
            break

        if new_distance < initial_distance:
            print(f"You're getting closer! New distance to treasure: {round(new_distance, 2)}")
        else:
            print(f"You're getting farther! New distance to treasure: {round(new_distance, 2)}")

    print("Current grid:")
    print_grid(player_position, treasure_position)

    initial_distance = new_distance
    moves_made += 1

    if moves_made == max_moves:
        print(f"Game over! You've run out of moves. The treasure was at {treasure_position}.")
        print_grid(player_position, treasure_position, show_treasure=True, hide_player=False)
```
